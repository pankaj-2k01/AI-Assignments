###Bucketizing the Column

```python
num_cols = data._get_numeric_data().columns
num_cols
```

```python
for col in num_cols:
  low = data[col].min()
  high = data[col].max()
  range = high - low + 1
  bins = [low, low + (1/3) * range, low +  (2/3) * range, high]
  labels = ['low', 'medium', 'high']
  data[col] = pd.cut(x = data[col], bins = bins, labels = labels, include_lowest = True)
```

```python
label_encoder = preprocessing.LabelEncoder() # to transform non-numerical labels to numerical ones
```

###Categorical Data

```python
cols = data.columns
num_cols = data._get_numeric_data().columns
categorical_labels=list(set(cols) - set(num_cols))
print(len(categorical_labels))
```

```python
categorical_labels
```

```python
for i in categorical_labels:
    data[i]= label_encoder.fit_transform(data[i])
```

###Splitting of the Dataset

```python
X=data.loc[:, data.columns != 'Suggested Job Role']
X
```

```python
Y=data['Suggested Job Role']
Y
```

# Train Test Split for 90-10 ratio

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=0)
```

###Logistic Regression

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs',max_iter=500)
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

###Random Forest

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

###SVM

```python
model = SVC(kernel='poly', degree=15)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

###XGB

```python
model = XGBClassifier()
model.fit(X_train, y_train)
xgb_y_pred = model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, xgb_y_pred))
```

###ANN MLP

```python
sc_X = StandardScaler()
X_trainscaled=sc_X.fit_transform(X_train)
X_testscaled=sc_X.transform(X_test)
model = MLPClassifier(activation='relu', solver = 'adam',early_stopping = True, hidden_layer_sizes=(128,64,32), verbose = 1, max_iter = 100)
model.fit(X_trainscaled, y_train)
y_pred=model.predict(X_testscaled)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
